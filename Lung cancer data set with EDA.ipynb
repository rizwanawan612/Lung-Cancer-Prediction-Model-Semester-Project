```python
def number_older(patient):
    Age, Gender= patient
    if Age > 60:
        return 'Older'
    else:
        return Gender
df['patient'] = df[['AGE','GENDER']].apply(number_older, axis=1)
```

```python
sns.countplot('COUGHING', data=df, hue='patient')
```

```python
df['AGE'].hist(bins=90)
```

```python
#Comparison of male, female, and older
df['patient'].value_counts()
```

```python
fig = sns.FacetGrid(df, hue="GENDER",aspect=5)
fig.map(sns.kdeplot,'AGE',shade= True)
fig.add_legend()
```

```python
#bandwidth effect can be seen here with increasing or decreasing the values
sns.kdeplot(df['AGE'], shade=True, bw=0.1) 
plt.show()
```
