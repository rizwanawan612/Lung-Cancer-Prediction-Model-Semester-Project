```python
#loading required packages
import pandas as pd
import matplotlib.pyplot as plt
#plt.style.use('classic')

%matplotlib inline
```

```python
# Data collect from https://data.world/cancerdatahp/lung-cancer-data and also this data on Kaggle
df=pd.read_csv("D:/survey lung cancer1.csv")
#df=data.copy()
df.head()
```

```python
df.columns
```

```python
# Aggregating population on year
df_popworld=df.groupby(['AGE'])['ALCOHOL CONSUMING'].sum().reset_index()
df_popworld.head()
```

```python
# Make a line plot: year on the x-axis, pop on the y-axis
age=df_popworld['AGE']
cholesterol=df_popworld['ALCOHOL CONSUMING']
plt.plot(age,cholesterol,"go--",color='cyan')
# Display the plot with plt.show() IT HELPS IN FLUSHING THE MEMORY
plt.show()
```
